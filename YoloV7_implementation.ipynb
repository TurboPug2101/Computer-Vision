```python
from moviepy.editor import VideoFileClip
from IPython.display import display, Video

video_path = '/content/yolov7/runs/detect/exp2/production_id_5198956 (2160p).mp4'
output_video_path = '/content/yolov7/runs/detect/exp2/production_id_5198956_reduced.mp4'

# Reduce video size
clip = VideoFileClip(video_path)
clip_resized = clip.resize(height=480)
clip_resized.write_videofile(output_video_path)

# Create a Video object with the reduced-size video file path
video = Video(output_video_path, embed=True)

# Display the video
display(video)
```
